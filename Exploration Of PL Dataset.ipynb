```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as  tkr
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
```

```python
#read data

data = pd.read_csv('complete_data.csv')
data.dropna(axis = 0, how = 'any', inplace = True)
data.info()
```

```python
data.columns = ['Final position','Team','Final points','Season','Wage','Wage scaled','Net transfer spend',
             'Net transfer spend scaled','Squad size','Avg age','Num foreign players', 'Team market value',
             'Is estimated market val','Team market value scaled','Avg time','Num transfers in',
             'Num transfers out','Stadium capacity scaled','Years in league', 'Big six']
data.head()
```

```python
data.drop(['Wage','Net transfer spend','Is estimated market val','Team market value'], axis=1, inplace=True)
```

```python
data.head()
```

```python
#Plot 1: Transfer Spend v Final points
sns.pairplot(data, vars = ['Net transfer spend scaled', 'Final points'], hue = 'Big six', height = 3.5, kind = 'reg')
```

```python
#Plot 2: Number of transfers in vs Final points
sns.pairplot(data, vars = ['Num transfers in', 'Final points'], hue = 'Big six', height = 3.5, kind = 'reg')
```
